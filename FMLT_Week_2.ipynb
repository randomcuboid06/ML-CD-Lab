# Measures of data spread

```python
import numpy as np
```

- quartiles

```python
quartiles = np.quantile(df['Max Temp (°C)'], [0, 0.25, 0.5, 0.75, 1])
```

- inter-quartile range

```python
print(quartiles[3] - quartiles[1])

# or

from scipy.stats import iqr

print(iqr(df['Max Temp (°C)']))
```

```python
df.describe()
```

- standard deviation

```python
hyd_df['Max Temp (°C)'].std()
```

- variance

```python
hyd_df['Max Temp (°C)'].var()
```
